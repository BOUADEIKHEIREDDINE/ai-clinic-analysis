```python
thermique= cons.groupby("Région").sum(numeric_only=True)['Thermique (MW)']
Nucleair = cons.groupby("Région").sum(numeric_only=True)['Nucléaire (MW)']
Eolien = cons.groupby("Région").sum(numeric_only=True)['Eolien (MW)']
solaire= cons.groupby("Région").sum(numeric_only=True)['Solaire (MW)']
consomation=cons.groupby("Région").sum(numeric_only=True)['Consommation (MW)']
hydrocarbures=cons.groupby("Région").sum(numeric_only=True)['Hydraulique (MW)']
Bioénergie=cons.groupby("Région").sum(numeric_only=True)['Bioénergies (MW)']
echange_ener=cons.groupby("Région").sum(numeric_only=True)['Ech. physiques (MW)']
```

```python
df1=pd.DataFrame()
```

### la consommation totale

```python
ProdTotal=thermique+Nucleair+Eolien+solaire+hydrocarbures+Bioénergie
ProdTotal
```

```python
consomation
```

```python
df1["Thermique"]=thermique
df1["Eolien"]=Eolien
df1["solaire"]=solaire
df1["consomation"]=consomation
df1["prodTotal"]=thermique+Nucleair+Eolien+solaire
df1["ehangeEner"]=echange_ener
df1["dispo"]=ProdTotal+echange_ener
df1["conso plus que dispo"]=df1.consomation>df1.dispo
```

```python
df1
```

### definition du heatmap des datas

```python
plt.figure(figsize=(20,10))
sb.heatmap(df1.corr(numeric_only=True),annot=True, cmap="coolwarm",fmt=".1g") 
plt.show()
```
