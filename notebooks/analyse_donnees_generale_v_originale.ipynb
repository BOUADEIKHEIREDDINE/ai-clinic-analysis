### projection des visualisations futures

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Charger les données
data = {
    'Région': [
        'Île-de-France', 'Normandie', 'Pays de la Loire', 'Bretagne',
        'Auvergne-Rhône-Alpes', 'Provence-Alpes-Côte d\'Azur'
    ],
    'Consommation (MW)': [
        1.194341e+11, 3.641567e+10, 3.118530e+10, 
        4.085306e+10, 1.244113e+11, 7.317488e+10
    ],
    'Nombre d\'entreprises (unités légales)': [
        908685, 140640, 166498, 157255, 
        454660, 363344
    ]
}
df = pd.DataFrame(data)

# Paramètres pour ajustements
transfer_percentage = 0.1  # Réduction dans les zones industrielles
industrial_redistribution = 0.05  # Augmentation d'industries dans les zones moins peuplées
renewable_increase = 0.08  # Gain d'énergie par les renouvelables dans les zones moins densément peuplées
high_density_regions = ['Île-de-France', 'Auvergne-Rhône-Alpes']
low_density_regions = ['Normandie', 'Bretagne']

# Simuler les ajustements
df['Consommation Projetée (MW)'] = df['Consommation (MW)']
df['Nombre d\'entreprises Projeté'] = df['Nombre d\'entreprises (unités légales)']

for region in high_density_regions:
    # Réduction dans les zones industrielles
    reduction = df.loc[df['Région'] == region, 'Consommation (MW)'].values[0] * transfer_percentage
    df.loc[df['Région'] == region, 'Consommation Projetée (MW)'] -= reduction
    # Redistribution de consommation et industries
    for low_region in low_density_regions:
        df.loc[df['Région'] == low_region, 'Consommation Projetée (MW)'] += reduction / len(low_density_regions)
        df.loc[df['Région'] == low_region, 'Nombre d\'entreprises Projeté'] += industrial_redistribution * df['Nombre d\'entreprises (unités légales)'].sum() / len(low_density_regions)

# Ajout de production locale renouvelable
for low_region in low_density_regions:
    renewable_gain = df.loc[df['Région'] == low_region, 'Consommation (MW)'].values[0] * renewable_increase
    df.loc[df['Région'] == low_region, 'Consommation Projetée (MW)'] -= renewable_gain

# Tracer la répartition
x = np.arange(len(df))
width = 0.35

plt.figure(figsize=(14, 7))
plt.bar(x - width/2, df['Consommation (MW)'], width, label="Consommation Actuelle (2023)", color="blue")
plt.bar(x + width/2, df['Consommation Projetée (MW)'], width, label="Consommation Projetée (2030)", color="green")

plt.xticks(x, df['Région'], rotation=45)
plt.ylabel("Consommation (MW)")
plt.title("Projection de la Consommation Énergétique (2023 vs 2030)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Affichage des données projetées
print(df)
```
